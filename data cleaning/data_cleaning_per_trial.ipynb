```python
import pandas as pd

# Step 1: Load Pre-study.csv
per_trial = pd.read_csv('/Users/Patron/Desktop/project/data/per-trial.csv')  
```

```python
per_trial.isna().sum()
```

```python
per_trial['input_type_encoded'] = per_trial['input_type'].map({
    'text': 0,
    'image': 1
})
```

```python
per_trial['model_order_encoded'] = per_trial['model_order'].map({
    'A, B': 0,
    'B, A': 1
})
```

```python
per_trial['diagnosis_type_encoded'] = per_trial['diagnosis_type'].map({
    'consistent': 0,
    'conflicting': 1,
    'partially consistent': 1
})
```

```python
per_trial['preferred_model'] = per_trial['preferred_model'].str.strip().str.capitalize()

per_trial['preferred_model_encoded'] = per_trial['preferred_model'].map({
    'A': 0,
    'B': 1,
    'Both': 2
})
```

```python
import re

def extract_confidence(value):
    if pd.isna(value):
        return 0
    value = value.strip()
    if value.lower() == 'no':
        return 0
    match = re.search(r'\((\d+)\)', value)
    if match:
        return int(match.group(1))
    else:
        return 0  # default if something weird happens

# Apply the function to both columns
per_trial['shown_confidence_A_numeric'] = per_trial['shown_confidence_A'].apply(extract_confidence)
per_trial['shown_confidence_B_numeric'] = per_trial['shown_confidence_B'].apply(extract_confidence)
```

```python
per_trial.drop(columns=['input_type','model_order','diagnosis_type','preferred_model', 'trust_reasoning_text','shown_confidence_A','shown_confidence_B'])
```
